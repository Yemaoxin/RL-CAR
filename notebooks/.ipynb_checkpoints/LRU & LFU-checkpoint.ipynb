```python
from environment import CacheEnv
from collections import defaultdict
```

```python
def lru_policy(s):
    least_recent = -1
    action = -1
    for key in s.keys():
        cur  = s[key][0]
        if  cur > least_recent:
            action = key
            least_recent = cur
        return action
```

```python
def lfu_policy(s):
    least_frequent = 100000000
    action = -1
    for key in s.keys():
        cur = s[key][1]
        if cur < least_frequent:
            action = key
            least_frequent = cur
        return action
```

```python
l = 20
trials = 50
results = defaultdict(list)

for l in range(10):
    for trial in range(trials):
        env = CacheEnv(eps_len=l)
        s = env.reset()
        done = env.done
#         print("Start: ", env.pages)
        while not done:
            a = lru_policy(s)
            s, r, done, observation = env.step(a)
        #     print(f">> Request: {env.new_page_id}")
#             print(f"Replace: {a}")
#             print(observation)
#             print(env.pages, f"reward: {r}\n")
        print(f"Total hits: {env.total_hits}")
        results[l].append(env.total_hits)
```
